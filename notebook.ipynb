```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("books.csv")
df.head(10)
```

```python

newDF = df[["average_rating",  "work_ratings_count"]]
HarryPotterBooks = df[["title"]]
HarryPotterBooks = HarryPotterBooks[HarryPotterBooks["title"].str.contains("Harry Potter")]
combinedDF = pd.concat([HarryPotterBooks, newDF], axis=1)
combinedDF = combinedDF.dropna()
combinedDF
```

```python
plt.bar(
    combinedDF["title"],
    combinedDF["work_ratings_count"],
    color="green",
)  
plt.xlabel("Title")
plt.ylabel("Ratings Count")
plt.title("Best-Selling Harry Potter Books")
plt.xticks(rotation=30, ha="right")  
plt.tight_layout()  
plt.show()
```

Average Rating of all Harry Potter books

```python
ratings = combinedDF["average_rating"]
avgRatings = round(numpy.mean(ratings),2)
avgRatings
```
